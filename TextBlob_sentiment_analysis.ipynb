```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer, PorterStemmer
from textblob import TextBlob
from gensim.parsing.preprocessing import remove_stopwords
import gensim
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer 
from sklearn.metrics import mean_squared_error 
```

```python
labeled_data = pd.read_csv("Unblinded AI Sentiment Mock Data & Scores.csv")
```

```python
labeled_data.head()
```

#### Stopwords removal (gensim)

```python
# change review type to string
labeled_data['Text'] = labeled_data['Text'].astype(str)
# lower case
labeled_data['Text'] = labeled_data['Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
## remove punctuation
labeled_data['Text'] = labeled_data['Text'].str.replace('[^\w\s]','')
```

```python
# remove stop words
all_stopwords = gensim.parsing.preprocessing.STOPWORDS
labeled_data['tidy_text'] = labeled_data['Text'].apply(lambda x: " ".join(x for x in x.split() if x not in all_stopwords))
```

```python
labeled_data.head()
```

#### Lemmatize Text

```python
lemmatizer = WordNetLemmatizer() 
labeled_data["tidy_text_lemmatized"] = labeled_data["tidy_text"].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
labeled_data.head(2)
```

#### Text stemming

```python
stemmerport = PorterStemmer()
labeled_data['tidy_text_stemmed'] = labeled_data["tidy_text"].apply(lambda x: ' '.join([stemmerport.stem(word) for word in x.split()]))
```

```python
labeled_data.head(2)
```

#### POS tagging

#### get prepositional/adverb words

```python
# adverb
def get_adverb(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "RB"]

labeled_data['RB'] = labeled_data['tidy_text_lemmatized'].apply(get_adverb)
```

```python
labeled_data.head()
```

```python
# preposition
def get_preposition(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "IN"]
labeled_data['IN'] = labeled_data['tidy_text_lemmatized'].apply(get_preposition)
```

```python
labeled_data.head(2)
```

```python
### Get adverb list
adverb_list = []
for index, row in labeled_data.iterrows():
    adverb_list.append(row["RB"])
```

```python
### Get preposition list
preposition_list = []
for index, row in labeled_data.iterrows():
    preposition_list.append(row["IN"])
```

```python
# adverb_list
```

```python
# flatten adverb list
adverb_list_flatten = [item for sublist in adverb_list for item in sublist]
# get unique
unique_adverb_list = list(set(adverb_list_flatten))
```

```python
unique_adverb_list
```

```python
# flatten preposition list
preposition_list_flatten = [item for sublist in preposition_list for item in sublist]
# get unique
unique_preposition_list = list(set(preposition_list_flatten))
```

```python
unique_preposition_list
```

```python
labeled_data["final_text"] = pd.DataFrame(labeled_data['tidy_text_lemmatized'].apply(lambda x: ' '.join([item for item in x.split() if ((item.lower() not in unique_adverb_list)  & (item.lower() not in unique_preposition_list))])))
```

```python
labeled_data.head(2)
```

#### TextBlob sentiment Analysis

```python
# the first score is sentiment polarity which tells if the sentiment is positive or negative and 
# the second score is subjectivity score to tell how subjective is the text.
def senti(x):
    return TextBlob(x).sentiment  
 
labeled_data['senti_score'] = labeled_data['final_text'].apply(senti)
 
labeled_data.senti_score.head()
```

```python
polarity_list = []
for index, row in labeled_data.iterrows():
    polarity_list.append(row["senti_score"].polarity)
```

```python
subjectivity_list = []
for index, row in labeled_data.iterrows():
    subjectivity_list.append(row["senti_score"].subjectivity)
```

```python
labeled_data["sentiment_score_textblob_new"] = polarity_list
```

```python
labeled_data["subjectivity_score_textblob_new"] = subjectivity_list
```

```python
labeled_data.head()
```

#### Report MSE

##### Improved mse

```python
y_true = list(labeled_data["Human Score"])
y_pred = list(labeled_data["sentiment_score_textblob_new"])
```

```python
mse_improved = mean_squared_error(y_true,y_pred)
```
